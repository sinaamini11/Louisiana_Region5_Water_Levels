```python
def read_data(folder):
    station = folder.split('=')[1]
    files = os.listdir(f'./temp/{folder}')
    file = list(filter(lambda x:x.endswith('.csv'), files))[0]
    temp = pd.read_csv(f'./temp/{folder}/{file}', low_memory=False)
    temp['DateTime'] = pd.to_datetime(temp['Date (mm/dd/yyyy)'] + ' ' + temp['Time (hh:mm:ss)']) 
    temp['Station ID'] = station
    temp['Lat (NAD83 - 2011)'] = r5_sta[r5_sta['STATION_ID']==station].iloc[0].loc['Lat']
    temp['Long (NAD83 - 2011)'] = r5_sta[r5_sta['STATION_ID']==station].iloc[0].loc['Long']
    temp['Basin'] = r5_sta[r5_sta['STATION_ID']==station].iloc[0].loc['BASIN']
    temp['Basin_1'] = r5_sta[r5_sta['STATION_ID']==station].iloc[0].loc['BASIN_1']
    outdf = temp[['Station ID', 'DateTime', 'Time Zone',
                  'Adjusted Water Elevation to Datum (ft)',
                  'Geoid', 'Lat (NAD83 - 2011)', 'Long (NAD83 - 2011)',
                  'Basin', 'Basin_1']].copy()
    outdf.sort_values('DateTime', inplace=True)

    return outdf
```

```python
import requests
def get_datum_shift(outdf):
    url = 'https://vdatum.noaa.gov/vdatumweb/api/tidal'
    response = requests.get(url, params={'lat':outdf['Lat (NAD83 - 2011)'].iloc[0],
                                         'lon':outdf['Long (NAD83 - 2011)'].iloc[0],
                                         's_v_frame':'NAVD88', 's_v_geoid':'geoid99',
                                         's_v_unit':'ft', 't_v_geoid':'geoid12b',
                                         't_v_unit':'ft', 't_v_frame':'NAVD88'})

    response = response.json() 
    return float(response['tar_height'])
```

```python
def process_data(raw_df):

    proc_df = raw_df.copy()

    assert (raw_df['Time Zone']=='CST').all()

    proc_df['DateTime'] = (raw_df['DateTime'] + pd.DateOffset(hours=-6))

    proc_df['Time Zone'] = 'GMT'

    shift = get_datum_shift(raw_df)

    proc_df['Adjusted Water Elevation to Datum (ft)'] = (raw_df['Adjusted Water Elevation to Datum (ft)']+
                                                        shift * (raw_df['Geoid']=='GEOID99'))

    proc_df['Geoid'] = proc_df['Geoid'].apply(lambda x: x.replace('GEOID99', 'GEOID12A'))

    assert (proc_df['Geoid'] == 'GEOID12A').all()

    return proc_df
```

```python
from bokeh.plotting import figure, output_file, show, save
from bokeh.models.tools import HoverTool

def plot(proc_df):

    TOOLS="xwheel_zoom,ywheel_zoom,crosshair,pan,reset,save"
    station = proc_df.iloc[0,0]
    p1 = figure(x_axis_type="datetime", title=f'{station}',
                tools=TOOLS, aspect_ratio=3, plot_width=1400,
                active_scroll='xwheel_zoom')

    p1.grid.grid_line_alpha=0.3
    p1.xaxis.axis_label = 'Date and Time (GMT)'
    p1.yaxis.axis_label = 'Water Level (ft-NAVD88 GEOID12B)'
    p1.line(x='DateTime', y='Adjusted Water Elevation to Datum (ft)', source=proc_df ,color='#4287f5')


    hover = HoverTool(tooltips=[('DateTime', '@DateTime{%Y-%m-%d %H:%M}'),
                                ('Water Level', "$y{%0.2f}")],
              formatters={'@DateTime': 'datetime', '$y':'printf'})

    p1.add_tools(hover)
    return p1
```

```python
def get_info(outdf):
    station = outdf['Station ID'].iloc[0]
    lat = outdf['Lat (NAD83 - 2011)'].iloc[0]
    long = outdf['Long (NAD83 - 2011)'].iloc[0]
    idxmax = outdf['Adjusted Water Elevation to Datum (ft)'].idxmax()
    idxmin = outdf['Adjusted Water Elevation to Datum (ft)'].idxmin()
    maxlevel = outdf['Adjusted Water Elevation to Datum (ft)'].max()
    minlevel = outdf['Adjusted Water Elevation to Datum (ft)'].min()
    max_date = str(outdf['DateTime'].loc[idxmax])
    min_date = str(outdf['DateTime'].loc[idxmin])
    repo = 'https://github.com/sinaamini11/Louisiana_Region5_Water_Levels'
    raw_file = f'{repo}/tree/main/temp/{folder}'
    proc_file = f'{repo}/blob/main/Output/{station}/{station}_Processed.csv'
    plot_file = f'https://htmlpreview.github.io/?{repo}/blob/main/Output/{station}/{station}.html'
    basin = outdf['Basin'].iloc[0]
    basin_1 = outdf['Basin_1'].iloc[0]
    return {'Station':station, 'Lat':lat, 'Long':long, 'Maxlevel':maxlevel, 
            'Maxdate':max_date, 'Minlevel': minlevel, 'Mindate':min_date, 
            'Raw': raw_file, 'Proc': proc_file, 'Plot':plot_file,
            'Basin':basin, 'Basin1':basin_1}
```

```python
def parallel_process(folder):
    if not os.path.isdir(f'./temp/{folder}'):
        return None
    station = folder.split('=')[1]
    raw_df = read_data(folder)
    if raw_df['Geoid'].isnull().all() or raw_df['Adjusted Water Elevation to Datum (ft)'].isnull().all():
        return None
    outdir = f'./Output/{station}'
    if not os.path.isdir(outdir):
        os.mkdir(outdir)
    outdf = process_data(raw_df)
    outdf.to_csv(f'{outdir}/{station}_Processed.csv', index=False)
    p = plot(outdf)
    output_file(f'{outdir}/{station}.html', mode='inline', title=station)
    save(p)
    return get_info(outdf)
    
```

```python
if not os.path.exists('./Output/map_data.csv'):
    from multiprocessing import Pool
    pool = Pool(8)
    folders = os.listdir('./temp')
    map_data = pool.map(parallel_process, folders)
    map_data = list(filter(lambda x:isinstance(x,dict), map_data))
    map_data = pd.DataFrame(map_data)
    map_data.to_csv('./Output/map_data.csv', index=False)
    pool.terminate()
else:
    map_data = pd.read_csv('./Output/map_data.csv')
```
